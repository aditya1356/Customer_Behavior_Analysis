```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(
    lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
#creat a column age_group
labels=['young adul','adult','middle-aged','senior']
df['age_groups'] =pd.qcut(df['age'], q=4, labels = labels )
```

```python
df[['age', 'age_groups']].head(10)
```

```python
# Corrected frequency mapping
frequency_mapping = {
    'fortnightly': 14,
    'weekly': 7,
    'monthly': 30,        # fixed spelling
    'quarterly': 90,      # lowercase for consistency
    'bi-weekly': 14,      # consistent casing
    'annually': 365,
    'every 3 months': 90  # fixed spelling
}

# Create new column
df['purchase_frequency_days'] = df['frequency_of_purchases'].str.lower().map(frequency_mapping)
```

```python
# Rename 'purchase_amount_(usd)' to 'purchase_amount'
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})

# Verify the change
print(df.columns)
```
